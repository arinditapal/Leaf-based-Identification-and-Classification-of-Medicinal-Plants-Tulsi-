```python
# changes the object type to float for glcm based features

def object_to_float(df, col):

    l = []
    for d in df[col]:
        value = float(d[3: len(d) - 1])
        l.append(value)

    df[col] = pd.DataFrame(l)

for i in range(8, 38):
    col = df.columns[i]
    object_to_float(df, col)

```

```python
df.head()
```

```python
# changing the label from object to numbers

le = LabelEncoder()

df["label"] = le.fit_transform(df["label"])

df.head()
```

```python
# stores the labels maap
labels = pd.DataFrame(le.inverse_transform(df["label"]))
labels
```

```python
top_25_features = ["homogen_4", "homogen_1", "homogen_5", "homogen_2", "homogen_3", "homogen_6", "dis_sim_6", "dis_sim_3", "energy_3", "energy_6", 'energy_2', 'energy_5', 'energy_4', 'energy_1', 'dis_sim_5', "dis_sim_2", 'dis_sim_1', "dis_sim_4", "contrast_6", "contrast_5", "perimeter_to_area_ratio", "contrast_3", "contrast_2", "contrast_4", 'contrast_1']

top_10_features = []
for i in range(10):
    top_10_features.append(top_25_features[i])


# print(len(top_25_features))

x = df.drop(["label", "image_name", "category"], axis=1)
y = df[["label"]]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=10)

# print(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators = 80, random_state = 5)

model.fit(x_train, y_train)
```

```python
prediction_test = model.predict(x_test)

print(prediction_test)
```

```python
from sklearn import metrics

print("Accuracy = ", metrics.accuracy_score(y_test, prediction_test))
```
